```python
# Mars Mission Challenge

# 🪐 Mars Mission Planner — Challenge Steps
#
# 1. Write a function to calculate how long it takes
#    to reach Mars at a given speed.
#    - Mars' average distance is 225,000,000 km.
#    - Use the formula: time = distance ÷ speed
#    - Round the result to the nearest hour.
#
# 2. Write another function to figure out the total fuel cost
#    for a mission.
#    - Formula: total cost = distance × fuel usage × price per liter
#
# 3. Test your functions with the provided mission data:
#    - Pathfinder: 40,000 km/h
#    - Perseverance: 75,000 km/h
#    - Starship: 120,000 km/h
#    - Each mission travels 225 million km,
#      burns 0.3 liters/km, and fuel costs $1.80/L.
#
# 4. For each mission, print a report showing:
#    - Mission name
#    - Travel time (hours)
#    - Total fuel cost

# Mission data
# Mission data
mission_1_speed = 40000  # Pathfinder
mission_2_speed = 75000  # Perseverance
mission_3_speed = 120000 # Starship

mars_distance = 225_000_000  # in kilometers
fuel_rate = 0.3              # liters per kilometer
fuel_price = 1.8             # dollars per liter

def travel_time(distance, speed):
    return round(distance / speed)

def fuel_cost(distance, rate, price):
    return distance * rate * price

# Pathfinder
path_time = travel_time(mars_distance, mission_1_speed)
path_cost = fuel_cost(mars_distance, fuel_rate, fuel_price)

per_time = travel_time(mars_distance, mission_2_speed)
per_cost = fuel_cost(mars_distance, fuel_rate, fuel_price)

star_time = travel_time(mars_distance, mission_3_speed)
star_cost = fuel_cost(mars_distance, fuel_rate, fuel_price)

print("Pathfinder:")
print("Travel time (hours):", path_time)
print("Total fuel cost ($):", path_cost)
print("\n Perseverance: ")
print("Travel_time: ", per_time)
print("Cost: $", per_cost)
print("\n Starship")
print("Total_time: ", star_time)
print("Total_cost: $", star_cost)
```

```python
# ☕ Coffee Order Queue Challenge
# 1. Set up two variables: one for total price, one for drink count
# 2. Start a while True loop
# 3. Ask for the customer's name
# 4. If the name is "done", break the loop
# 5. Ask for their drink order
# 6. If it's "latte", add 3.50 to total and +1 to drink count
#    If it's "americano", add 3.00 to total and +1 to drink count
#    If it's "espresso", add 2.50 to total and +1 to drink count
# 7. If it's not one of those drinks, print a warning and continue
# 8. After the loop, print total number of drinks and total price

total_price = 0
drink_count = 0
customer_name = input('What is your name: ')
while True:
    drink_order = input("What is your order latte, americano, espresso or 'done' to finish: ")
    if drink_order == "done":
        break
    elif drink_order == "latte":
        total_price += 1.35
        drink_count += 1
    elif drink_order == "americano":
        total_price += 1.00
        drink_count += 1
    elif drink_order == "espresso":
        total_price += 2.30
        drink_count += 1
    else:
        print("Sorry we dont serve that here")
        continue

print(f"Total Drinks = {drink_count}")
print(f"Total Bill = {total_price}")
```

```python
# 📱 Phone Number Formatter
#
# 1. Ask the user to enter a U.S. phone number in **any format**.
# 2. Use .strip() to remove any leading/trailing spaces.
# 3. Replace common separators (-, (, ), .) with spaces.
# 4. Use .split() to break into chunks, then .join() to merge the digits.
# 5. Check if the cleaned number has **exactly 10 digits**.
# 6. If yes, format it like this: (123) 456-7890
# 7. If not, print an error message: "Please enter exactly 10 digits."

phone_input = input("Enter a U.S. phone number in any format: ")

cleaned_input = phone_input.strip()  # Remove leading/trailing spaces

for char in ['-', '(', ')', '.']:
    cleaned_input = cleaned_input.replace(char, ' ')  # Replace separators with spaces

digits = cleaned_input.split()  # Split into chunks
merged_digits = ''.join(digits)  # Merge the chunks

if len(merged_digits) == 10:
    formatted_number = f"({merged_digits[:3]}) {merged_digits[3:6]}-{merged_digits[6:]}"
    print("Formatted phone number:", formatted_number)
else:
    print("Error: Please enter exactly 10 digits.")
```

```python
# 🛂 Access Control Scanner Challenge
#
# 1. Create a set of revoked badge numbers.
# 2. Create two empty lists: "approved" and "denied".
# 3. Start a loop to collect visitor info:
#    - Ask for the visitor's name (or type "done" to finish).
#    - If the name is "done", exit the loop.
#    - Otherwise, ask for their badge number.
#    - Check if the badge is revoked:
#        • If revoked: add the name to "denied" and display "ACCESS DENIED".
#        • If not: add the name to "approved" and display "ACCESS GRANTED".
# 4. Print the final "Access Summary" for "✅ Approved Visitors" & "⛔️ Denied Visitors":
#    - Sort both lists alphabetically.
#    - Display the total number of approved and denied visitors.

rev_badge = {12345, 67890, 24680, 13579}  # Set of revoked badge numbers
approved = []  # List to store approved visitors
denied = []    # List to store denied visitors
while True: # starts an infinite loop
    name = input("Enter visitor's name (or 'done' to finish): ") # takes input for the visitor's name
    if name.lower() == "done": # checks if the name is 'done'
        break
    badge = int(input("Enter badge number: ")) # takes input for the badge number
    if badge in rev_badge: # checks if the badge number is in the set of revoked badges
        denied.append(name) # adds the name to the list of denied visitors
        print("ACCESS DENIED") # prints 'ACCESS DENIED'
    else:
        approved.append(name) # adds the name to the list of approved visitors
        print("ACCESS GRANTED") # prints 'ACCESS GRANTED'
# Print final access summary
print("\n--- Access Summary ---") 
print("✅ Approved Visitors:")
for visitor in sorted(approved): # iterates over the sorted list of approved visitors
    print(f" - {visitor}") # prints each approved visitor
print("⛔️ Denied Visitors:")
for visitor in sorted(denied): # iterates over the sorted list of denied visitors
    print(f" - {visitor}") # prints each denied visitor
print(f"Total Approved: {len(approved)}") # prints the total number of approved visitors
print(f"Total Denied: {len(denied)}")
```
